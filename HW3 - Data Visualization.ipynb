```python
df_raw = pd.DataFrame()

for i in range(n):
    data = pd.read_csv(files_csv[i])
    data['Year'] = int(files_csv[i][0:4])
    df_raw = df_raw.append(data)

df_raw = df_raw.fillna(0)
```

# Задание 3
## Визуализировать эту информацию: Вывести распределения показателей счастья (Health, Freedom, Trust и т.д.) в том или ином виде, используя Seaborn


```python
df_3 = pd.DataFrame()
df_3['Country'] = df_raw['Country']

df_3['Happiness_Score_2015'] = df_raw['Happiness Score'].where((df_raw['Year'] == 2015))
df_3['Happiness_Score_2016'] = df_raw['Happiness Score'].where((df_raw['Year'] == 2016))
df_3['Happiness_Score_2017'] = df_raw['Happiness Score'].where((df_raw['Year'] == 2017))

df_3['Health_Score_2015'] = df_raw['Health (Life Expectancy)'].where((df_raw['Year'] == 2015))
df_3['Health_Score_2015'] = df_raw['Health (Life Expectancy)'].where((df_raw['Year'] == 2016))
df_3['Health_Score_2015'] = df_raw['Health (Life Expectancy)'].where((df_raw['Year'] == 2017))

df_3['Freedom_Score_2015'] = df_raw['Freedom'].where((df_raw['Year'] == 2015))
df_3['Freedom_Score_2015'] = df_raw['Freedom'].where((df_raw['Year'] == 2016))
df_3['Freedom_Score_2015'] = df_raw['Freedom'].where((df_raw['Year'] == 2017))

df_3['Trust_Score_2015'] = df_raw['Trust (Government Corruption)'].where((df_raw['Year'] == 2015))
df_3['Trust_Score_2015'] = df_raw['Trust (Government Corruption)'].where((df_raw['Year'] == 2016))
df_3['Trust_Score_2015'] = df_raw['Trust (Government Corruption)'].where((df_raw['Year'] == 2017))

df_3 = df_3.fillna(0)

df_3 = df_3.groupby('Country').sum()
```
