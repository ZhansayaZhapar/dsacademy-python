```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

** Import matplotlib.pyplot as plt and set %matplotlib inline if you are using the jupyter notebook. What command do you use if you aren't using the jupyter notebook?**

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Exercise 1

** Follow along with these steps: **
* ** Create a figure object called fig using plt.figure() **
* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("title")

ax.set_ylim([0, 200])
ax.set_xlim([0, 100])

ax.set_yticks([0, 50, 100, 150, 200])

ax.tick_params(axis="y",direction="in", pad=15)
ax.tick_params(axis="x",direction="in", pad=15)


ax.plot(x, y, 'b')
```

** Now plot (x,y) on both axes. And call your figure object to show it.**

```python
fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1]) 
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2]) 

ax1.plot(x, y, 'darkred')
ax1.xaxis.set_ticks(np.arange(0, 100, 20))
ax1.yaxis.set_ticks(np.arange(0, 200, 25))

ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_ylim([0, 200])
ax1.set_xlim([0, 100])
ax1.set_yticks([0, 50, 100, 150, 200])



ax2.plot(x, y, 'darkred')
ax2.xaxis.set_ticks(np.arange(0, 100, 100))
ax2.yaxis.set_ticks(np.arange(0, 200, 200))

ax2.set_xlabel("x")
ax2.set_ylabel("y")

ax2.set_ylim([0, 200])
ax2.set_xlim([0, 100])

ax2.set_xticks([0, 20, 40, 60, 80, 100])
ax2.set_yticks([0, 50, 100, 150, 200])
```

** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
fig, axes = plt.subplots(2, 1, figsize=(8,12))
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1]) 
ax2 = fig.add_axes([0.2,0.5,.4,.4]) 

ax1.plot(x, z, 'b')

ax1.set_xlabel("X")
ax1.set_ylabel("Z")
ax1.set_ylim([0, 10000])
ax1.set_xlim([0, 100])
ax1.set_yticks([0, 2000, 4000, 6000, 8000, 10000])

ax2.plot(x, y, 'b')

ax2.set_xlabel("X")
ax2.set_ylabel("Y")

ax2.set_ylim([30, 50])
ax2.set_xlim([20, 22])

ax2.set_xticks([20.0, 20.5, 21.0, 21.5, 22.0])
ax2.set_yticks([30, 35, 40, 45, 50])
ax2.set_title('zoom')
```

```python
#from mpl_toolkits.axes_grid1.inset_locator import inset_axes

fig, axes = plt.subplots(2, 1, figsize=(8,12))
    
for i in range(2):
    axes[i].plot(x, z, 'b')
   
    axes[i].set_xlabel("X")
    axes[i].set_ylabel("Z")
    axes[i].set_ylim([0, 10000])
    axes[i].set_xlim([0, 100])
    axes[i].set_yticks([0, 2000, 4000, 6000, 8000, 10000])
    
    inset_ax = axes[i].inset_axes(([0.2,0.5,.4,.4]))
 
    inset_ax.set_ylim([30, 50])
    inset_ax.set_xlim([20.0, 22.0])
    inset_ax.plot(x, y, 'b')
    
    inset_ax.set_xlabel("X")
    inset_ax.set_ylabel("Y")
```
