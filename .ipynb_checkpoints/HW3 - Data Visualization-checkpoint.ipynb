```python
path = os.getcwd()
files = os.listdir(path)
files_csv = [f for f in files if f[-3:] == 'csv']
files_csv
```

```python
n = len(files_csv)
```

```python
n
```

```python
df_raw = pd.DataFrame()

for i in range(n):
    data = pd.read_csv(files_csv[i])
    data['Year'] = int(files_csv[i][0:4])
    df_raw = df_raw.append(data)

df_raw = df_raw.fillna(0)
```
